```python
#pandas is a library.
#import refers to load the function in library.
#pd refers as short form of pandas
import pandas as pd
```

```python
#here the dataset is loaded via csv file
dataset=pd.read_csv("salary_data.csv")
```

```python
#calling dataset
dataset
```

```python
#picking up input variable in dataset
independent=dataset[["YearsExperience"]]
```

```python
#calling independent
independent
```

```python
#picking up output variable in dataset
dependent=dataset[["Salary"]]
```

```python
#calling output
dependent
```

```python
#from sklearn library model is selected imported the train,test,split
from sklearn.model_selection import train_test_split
#assuming test size=0.30(1/3rd of the total size)
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#by using linear regression we create the model
#linear regression is considered as class
from sklearn.linear_model import LinearRegression
#creating variable for linear regression
regressor=LinearRegression()
#calling regressor in fit function via access operator
#x-train = input, y-train = output respectively
regressor.fit(X_train,Y_train)
```

```python
#to wight is calculated using coef_
weight=regressor.coef_
weight
```

```python
#calculating bais
bais=regressor.intercept_
bais
```

```python
#evalumetric test for prediction
#End goal of AI is pediction
Y_pred=regressor.predict(X_test)
```

```python
#the tested score calculated via r_score
#nearly to 1 means good model(good performance)
#otherwise said to be bad model(poor performance)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
#nearly to 1  so, it is good model
r_score
```

```python
#pickle is used to save model
import pickle
#create filename pickle extenion save = sav
filename="finalized_model_linear.sav"
```

```python
#open file name in writing in binary(wb)
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the model rb is to read the model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#here the prediction is done
result=loaded_model.predict([[13]])
```
